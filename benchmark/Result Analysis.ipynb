# Classificazione

```python
df_clf = pd.read_csv('../results/CLF/results.csv')

df_clf.shape
```

```python
print(f"Dataset[{len(df_clf.dataset.unique().tolist())}]: {df_clf.dataset.unique().tolist()}", "\r\n")
print(f"Models[{len(df_clf.model.unique())}]: {df_clf.model.unique()}", "\r\n")
```

```python
df_clf.columns
```

```python
validation = [x for x in df_clf.columns if "validation_" in x]
test = [x for x in df_clf.columns if "test_" in x]
other_measures = ["fit_time"]
hyperparams = list(set(df_clf.columns) - set(validation) - set(test) - set(other_measures))
```

```python
df_clf = df_clf.groupby(by=list(set(hyperparams) - {"random_state"}), dropna=False).mean().reset_index()

print("#experiments", len(df_clf))
```

```python
df_grouped = df_clf.groupby(by=["dataset", "model"]).max().reset_index()

df_joined = df_grouped[df_grouped.model == "DT"].merge(df_grouped[df_grouped.model == "RT"], on=["dataset"], suffixes=("_DT", "_RT"))

plt.scatter(df_joined.dataset, df_joined.validation_accuracy_DT, label="DT")
plt.scatter(df_joined.dataset, df_joined.validation_accuracy_RT, label="RT")

plt.xticks(rotation=90)
plt.ylim(.5, 1)
plt.legend()

plt.show()
```

```python
plt.scatter(df_joined.dataset, df_joined.validation_accuracy_RT - df_joined.validation_accuracy_DT, label="DT")

max_val_y = max(np.abs(list(plt.ylim())))
plt.ylim(-max_val_y, max_val_y)
plt.xticks(rotation=90)
plt.hlines(0, *plt.xlim())

plt.figtext(.2, .8, 'RT win')
plt.figtext(.2, .2, 'DT win')

plt.show()
```

```python
plt.scatter(df_joined.dataset, df_joined.fit_time_DT - df_joined.fit_time_RT, label="DT")

max_val_y = max(np.abs(list(plt.ylim())))
plt.ylim(-max_val_y, max_val_y)
plt.xticks(rotation=90)
plt.hlines(0, *plt.xlim())

plt.figtext(.2, .8, 'RT win')
plt.figtext(.2, .2, 'DT win')

plt.show()
```

```python
group_time_cols = [
    "dataset",
    "min_weight_fraction_leaf", 'oob_score', 'class_weight', 'warm_start', 'max_n_vals_cat', 'task', 'splitter', 'max_features',
    'min_samples_leaf', 'ccp_alpha', 'max_depth', 'min_samples_split', 'max_leaf_nodes', 'max_features', 'criterion', 'max_n_vals', 'max_n_vals_cat'
]

df_grouped_time = df_clf[df_clf.model == "DT"].merge(df_clf[df_clf.model == "RT"], on=group_time_cols, suffixes=("_DT", "_RT"))
```

```python
df_grouped_time[[x for x in df_grouped_time.columns if x[:-3] not in validation+test+group_time_cols]].drop(columns=[
    "random_state_DT", "random_state_RT", "n_jobs_DT", "bootstrap_DT", "n_jobs_RT", "bootstrap_RT", "max_samples_RT", "max_samples_DT",
    "n_estimators_DT", "n_estimators_RT", "prune_useless_leaves_DT", "model_RT", "model_DT"
])
```

```python
df_grouped_time.groupby(by=["one_hot_encode_cat_RT", "max_depth"], dropna=False)[["fit_time_RT", "fit_time_DT"]].mean()
```

# Regressione

```python
df_reg = pd.read_csv('../results/REG/results.csv')

df_reg.shape
```

```python
print(f"Dataset[{len(df_reg.dataset.unique().tolist())}]: {df_reg.dataset.unique().tolist()}", "\r\n")
print(f"Models[{len(df_reg.model.unique())}]: {df_reg.model.unique()}", "\r\n")
```

```python
df_reg.columns
```

```python
validation = set([x for x in df_reg.columns if "validation_" in x])
test = set([x for x in df_reg.columns if "test_" in x])
other_measures = ["fit_time"]
hyperparams = set(df_clf.columns) - set(validation) - set(test) - other_measures
```

```python
df_reg = df_reg.groupby(by=list(hyperparams), dropna=False).mean().reset_index()

print("#experiments", len(df_reg))
```

```python
df_grouped = df_reg.groupby(by=["dataset", "model"]).max().reset_index()

plt.scatter(df_grouped[df_grouped.model == "DT"].dataset, df_grouped[df_grouped.model == "DT"].validation_r2, label="DT")
plt.scatter(df_grouped[df_grouped.model == "RT"].dataset, df_grouped[df_grouped.model == "RT"].validation_r2, label="RT")
plt.scatter(df_grouped[df_grouped.model == "RF"].dataset, df_grouped[df_grouped.model == "RF"].validation_r2, label="RF")
plt.scatter(df_grouped[df_grouped.model == "RTF"].dataset, df_grouped[df_grouped.model == "RTF"].validation_r2, label="RTF")

plt.xticks(rotation=90)
plt.legend()

plt.show()
```

```python
df_joined_tree = df_grouped[df_grouped.model == "DT"].merge(df_grouped[df_grouped.model == "RT"], on=["dataset"], suffixes=("_DT", "_RT"))
df_joined_forest = df_grouped[df_grouped.model == "RF"].merge(df_grouped[df_grouped.model == "RTF"], on=["dataset"], suffixes=("_RF", "_RTF"))

fig, ax1 = plt.subplots()

lns1 = ax1.scatter(df_joined_tree.dataset, 
            df_joined_tree.validation_r2_RT - df_joined_tree.validation_r2_DT, 
            label="Decision Trees")

ax2 = ax1.twinx()

lns2  = ax2.scatter(df_joined_forest.dataset, 
            df_joined_forest.validation_r2_RTF - df_joined_forest.validation_r2_RF, 
            label="Random Forests", c="tab:orange")

max_val_y = max(np.abs(list(ax1.get_ylim())))
ax1.set_ylim(-max_val_y, max_val_y)
max_val_y = max(np.abs(list(ax2.get_ylim())))
ax2.set_ylim(-max_val_y, max_val_y)

plt.hlines(0, *plt.xlim())

lns = [lns1, lns2]
labs = [l.get_label() for l in lns]
ax1.legend(lns, labs, loc=0)

plt.setp(ax1.get_xticklabels(), rotation=90, ha='right')
ax1.set_ylabel(r"R2 Decision Tree")
ax2.set_ylabel(r"R2 Random Forest")

plt.figtext(.2, .8, 'RT win')
plt.figtext(.2, .2, 'DT win')

plt.show()
```

```python
validation = set([x for x in df_clu.columns if "validation_" in x])
test = set([x for x in df_clu.columns if "test_" in x])
hyperparams = set(df_clu.columns) - validation - test
```

```python
df_clu = df_clu.groupby(by=list(hyperparams), dropna=False).mean().reset_index()

print("#experiments", len(df_clu))
```

```python
df_grouped = df_clu.groupby(by=["dataset", "model"]).max().reset_index()

plt.scatter(df_grouped[df_grouped.model == "KM"].dataset, df_grouped[df_grouped.model == "KM"].train_silhouette_score, label="KM")
plt.scatter(df_grouped[df_grouped.model == "KT"].dataset, df_grouped[df_grouped.model == "KT"].train_silhouette_score, label="KT")
#plt.scatter(df_grouped[df_grouped.model == "RT"].dataset, df_grouped[df_grouped.model == "RT"].train_silhouette_score, label="RT")

plt.xticks(rotation=90)
plt.legend()

plt.show()
```

```python
df_joined_kmeans = df_grouped[df_grouped.model == "KM"].merge(df_grouped[df_grouped.model == "RT"], on=["dataset"], suffixes=("_KM", "_RT"))
df_joined_kdtree = df_grouped[df_grouped.model == "KT"].merge(df_grouped[df_grouped.model == "RT"], on=["dataset"], suffixes=("_KT", "_RT"))

fig, ax1 = plt.subplots()

lns1 = ax1.scatter(df_joined_kmeans.dataset, 
            df_joined_kmeans.train_silhouette_score_RT - df_joined_kmeans.train_silhouette_score_KM, 
            label="RuleTree - kMeans")

ax2 = ax1.twinx()

lns2  = ax2.scatter(df_joined_kdtree.dataset, 
            df_joined_kdtree.train_silhouette_score_RT - df_joined_kdtree.train_silhouette_score_KT, 
            label="RuleTree - kMeansTree", c="tab:orange")

max_val_y = max(np.abs(list(ax1.get_ylim())))
ax1.set_ylim(-max_val_y, max_val_y)
max_val_y = max(np.abs(list(ax2.get_ylim())))
ax2.set_ylim(-max_val_y, max_val_y)

plt.hlines(0, *plt.xlim())

lns = [lns1, lns2]
labs = [l.get_label() for l in lns]
ax1.legend(lns, labs, loc=0)

plt.setp(ax1.get_xticklabels(), rotation=90, ha='right')
ax1.set_ylabel(r"R2 Decision Tree")
ax2.set_ylabel(r"R2 Random Forest")

plt.figtext(.2, .8, 'RT win')
plt.figtext(.2, .2, 'DT win')

plt.show()
```
