```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("fair_benchmark.csv")

df.head()
```

```python
df.columns
```

```python
col_viz = ["base_method", "method", "max_depth", "n_clusters", "use_t", "silhouette_score", "fowlkes_mallows_score", "balance_metric", "max_fairness_cost"]

df.sort_values(by=["balance_metric"], ascending=False).groupby(by=["base_method", "method", "use_t"], dropna=False).head(1)[col_viz].sort_values(by=col_viz)
```

```python
df.sort_values(by=["max_fairness_cost"], ascending=True).groupby(by=["base_method", "method", "use_t"], dropna=False).head(1)[col_viz].sort_values(by=col_viz)
```

```python
df.sort_values(by=["silhouette_score"], ascending=False).groupby(by=["base_method", "method", "use_t"], dropna=False).head(1)[col_viz].sort_values(by=col_viz)
```
